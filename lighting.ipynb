```python
import pandas as pd
import plotly_express as px
import plotly
```

```python
file_path = '2022-2023.csv'
```

```python
data_df = pd.read_csv(file_path)
data_df.head()
```

```python
data_df.info()
```

```python
data_df['CRASH_DATE'] = pd.to_datetime(data_df['CRASH_DATE'], format= '%m/%d/%Y %H:%M' )
data_df.head()
```

```python
data_df.info()
```

```python
data_df_sorted = data_df.sort_values(by='CRASH_DATE')
data_df_sorted.tail(100)
```

```python
#columns_to_choose = ['CRASH_DATE', 'LIGHTING_CONDITION', 'INJURIES_TOTAL' ]

##new_df.head()
```

```python
lighting = data_df_sorted['LIGHTING_CONDITION']
date = data_df_sorted['CRASH_DATE'].dt.strftime('%Y-%m-%d')
#num_injuries = new_df['INJURIES_TOTAL']

#print(num_injuries)


```

```python
lighting_counts = data_df_sorted['LIGHTING_CONDITION'].value_counts().reset_index()
lighting_counts.columns = ['LIGHTING_CONDITION', 'Count']
print(lighting_counts)
```

```python
fig = px.bar(
    lighting_counts,
    x = 'LIGHTING_CONDITION',
    y = 'Count',
    color='LIGHTING_CONDITION',
    title = 'Crashes by lighting Condition'

)


dropdown_options = [
    {'label': lighting, 'method': 'update', 'args': [{'visible': lighting_counts['LIGHTING_CONDITION'] == lighting}]}
    for lighting in lighting_counts['LIGHTING_CONDITION']
]

# Add an option to show all data
dropdown_options.append({'label': 'Show All', 'method': 'update', 'args': [{'visible': True}]})

# Create dropdown menu
fig.update_layout(
    updatemenus=[
        {
            'buttons': dropdown_options,
            'direction': 'down',
            'showactive': True,
            'x': 0.1,
            'xanchor': 'left',
            'y': 1.1,
            'yanchor': 'top'
        },
    ]
)



plotly.offline.plot(fig, filename='lighting.html')
```

```python
weather_counts = data_df_sorted['WEATHER_CONDITION'].value_counts().reset_index()
weather_counts.columns = ['WEATHER_CONDITION', 'Count1']
print(weather_counts)
```

```python
fig2 = px.bar(
    weather_counts,
    x = 'WEATHER_CONDITION',
    y = 'Count1',
    color='WEATHER_CONDITION',
    title = 'Crashes by Weather Condition'

)

buttons = [
    dict(label="All",
         method="update",
         args=[{"visible": [True] * len(weather_counts)}])
]
buttons += [
    dict(label=weather_condition,
         method="update",
         args=[{"visible": [weather == weather_condition for weather in weather_counts['WEATHER_CONDITION']]}])
    for weather_condition in weather_counts['WEATHER_CONDITION']
]

# Add dropdown menu
fig2.update_layout(
    updatemenus=[
        dict(
            buttons=buttons,
            direction="down",
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

plotly.offline.plot(fig2, filename='weather.html')
```
